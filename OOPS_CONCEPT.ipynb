```
import text2emotion as te  #To know about the emotion of Chitti 
import random  #To randomly select the action of chitti
```

```
# The class that will help inbuiding the replicas of Chitti
class BuildRep():
  replicas_list=['Building_Construction','House_Hold_Work','Carpenter','Painter','Music Composer']

  #The method that will list differnt format of replicas build by Chitti
  def Types_of_Replica(self):
    print("\nThe list of replicas that I can build are:-\n")
    for i in BuildRep.replicas_list:
      print(i)
    print("\n\n")

  #The method that will list the functionalities to be held by the Humanoid Robot
  def functionalities(self):
    print('@'*40,"\n\tTHE TYPES OF REPLICAS \t\n",'@'*40)
    self.Types_of_Replica()
    funct='''   **********************************************************************
    **********************************************************************
    The Humanoid Robot will be built on the follwing requirements:-
    @..1) Intelligence
    @..2) Sense
    @..3) Dexterity(Functions of limbs, appendages and extremities
    @..4) Power
    @..5) Independence
    **********************************************************************
    **********************************************************************
    '''
    print(funct,"\n\n")

  #The method that will activate the building of replica
  def replicabuilding(self):
    #To list the details of the Robot functionalities and types of replicas that can be built
    self.functionalities() 
    repli_choice=random.choice(BuildRep.replicas_list)
    print("\tPresntly I am building",repli_choice," Humanoid Robot. Its features are\n")
    if repli_choice==BuildRep.replicas_list[0]:
      print("\tThis Robot will help in construcction of building.\n\tIt can engineer the building process and can also work on fields.\n\tIt can hold the qualities of Civil Engineer and masion\n")
    elif repli_choice==BuildRep.replicas_list[1]:
      print("\tThis Robot can do all sort of household work.\n\tIt can clean the floor and walls.\n\tSet the infrasstructure in room.\n\tIt can even Cook\n")
    elif repli_choice==BuildRep.replicas_list[2]:
      print("\tThis Robot can differentiate the automobile parts.\n\tRepair all types of vehicles\n\tPaint the vehicle\n")
    elif repli_choice==BuildRep.replicas_list[3]:
      print("\tThis Robot can paint the walls\n\tIt can also do digital painting and design the brochures\n\tIt can teach the students gow to paint\n")
    else:
      print("\tThis Robot will help in composing the Music.\n\tIt will tune lyrics\n\tMarket the tonned music thoughtout the Globe\n")

    print("\tThe robot is build based on the analysed data that was fed to me earlier.\n")
```

```
#This class will help Chitti in finding out the emotions of User
class Emo_Des():
  InpStr=""
  UserEmotion=''
  #Method that will interact with the user
  def InteractUser(self):
    print('@'*40,"\n\t HELLO SIR!\n\tHOW ARE YOU DOING?\n",'@'*40)
    
    
    try:
      Emo_Des.InpStr=input("\t")
      if Emo_Des.InpStr=="":
        raise Exception

    except Exception:
      print("DO ENTER WHAT DO YOU FEEL NOW")

    Emo_Des.UserEmotion=te.get_emotion(Emo_Des.InpStr)
    print('*'*40,"\nYour Current Emotional scale values are:-\n\t\n",'*'*40)
    print(Emo_Des.UserEmotion)
    print("\n\t-----HOPE THE SCALING OF YOU EMOTIONS ARE USEFUL-----\n")
```

```
#This class will describe the details of Chitti
class Chitti(BuildRep,Emo_Des):
  speed="1GHz"
  memory="1TB"
  emotion=''

  #Method that describes it characters
  def Des_char(self):
    print('-'*40,"\n Hello, This is Chitti... The Humanoid Robot\n")
    print('-'*40 ,"\n\tSpeed = 1 Tera Hertz \n\tMemory= 1 Tera Bytes")
    print('-'*40,"\n\n\n")
  

  #Reading and analyzing the human emotions
  def humanemotions(self):
      try:
        cho=input("Would you like to know Your Emotional Status(YES/NO):\n")
        if cho=='YES':
          super().InteractUser()
        elif cho=='NO':
          print('*'*40,"\n---Signing off---\n",'*'*40)
        else:
          raise TypeError
      
      except TypeError:
        print("Sorry, Invalid Choice\n Enter either YES/NO\n")
        self.humanemotions()


  #Method to read the emotions of Chitti
  def Feelings(self):
    ModeSwings=['My battery is low','Full of Power','Having great fun building my duplicates','Power off Mood','Missing something']
    cm=random.choice(ModeSwings)
    Chitti.emotion=te.get_emotion(cm)
    print('-'*80,"\n\tThe current mode of Chitti\n\t",cm,"\n\t",Chitti.emotion,"\n",'-'*80,"\n\n\n")
    

    #Verfying the mode of the Chitti. And switch it activities
    try:
      if cm=="My battery is low":
        raise Exception
      elif cm=="Full of Power":
        raise Exception
      elif cm=="Having great fun building my duplicates":
        print ("Great to know that...\nContinue buliding the replicas...")
        replicabuilding()
      elif cm=="Power off Mood":
        print('-'*40,"Charge yourself....")
        raise Exception
      else:
        raise Exception
      
    except Exception:
        print("Dear Chitti, is not in the mode to build constructive Robot.\nIt's sad and angry levels are fluctuating.\nIt's better to pause the building process\n", '-'*80,"\n\n\n")
        self.humanemotions()

    
```

```
#The main class that will help in Dr.Vasi Interacting with Chitti.
class Vasi(Chitti):
  Name="Dr.Vasi"
  Password="Chitti_Bro" 

  def __init__(self):
    print("\n\n\n\tWELCOME\n\n\n")

  def verification(self):
    try:
      name=input("May, I know who is interactiong with me:\n")
      if name!=Vasi.Name:
        raise ValueError
      password=input("Enter Your Password:\n")
      if password!=Vasi.Password:
        raise ValueError
      
    except ValueError:
      print("\n\nYour credentials are wrong.\nTry entering correct UserName/Password\n")
      self.verification()

  def Interface(self):
    self.verification()
    print("\n\n\n")
    print('@'*40,"\n\tHELLO Dr.Vasi\n\tMy Desginer and Brother\n",'@'*40,"\n\n")
    super().Des_char()
    print("\tChitti, What is ur present working status!!!\n\n")
    super().Feelings()
```
